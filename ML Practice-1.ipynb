**For Object type data Pre-Processing**





```python
data['Awards'] = data['Awards'].fillna(data['Awards'].mode()[0])
data['Released'] = data['Released'].fillna(data['Released'].mode()[0])
data['Writer'] = data['Writer'].fillna(data['Writer'].mode()[0])
data['DVD'] = data['DVD'].fillna(data['DVD'].mode()[0])
data['BoxOffice'] = data['BoxOffice'].fillna(data['BoxOffice'].mode()[0])
data['Website'] = data['Website'].fillna(data['Website'].mode()[0])
```
